## **iris dataset**:  
data : 150 x 4 array    
feature_names : 4 array. iris.data is 150ea sample data of 4 features    
target : Speicies in number   
target_names : Name of each Target

#### **Make DataFrame**

```python
df_iris = pd.DataFrame(columns = iris.feature_names)
```

```python
for idx, col in enumerate(df_iris.columns):
    df_iris[col] = iris.data[:,idx]
```

```python
df_iris.head()
```

#### **View dataset with Seaborn**

```python
df_iris['species'] = iris.target
```

```python
import seaborn as sns
sns.scatterplot(data = df_iris, x = 'sepal length (cm)', y = 'sepal width (cm)', hue = 'species')
```

## Machine Learning   
Split df to train and test by using sklearn train_test_split module    
And classify test data with RFC   
RFC is contatined in sklearn module, we can use it easily

```python
from sklearn.model_selection import train_test_split
```

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
x = df_iris[features]
y = df_iris['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(
    n_estimators = 100,
    max_depth = 5,
    random_state = 1    
)
```

```python
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

### Validation

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average = 'weighted')
print(f'Accuracy : {acc_score} \nf1_score = {f1}')
print(classification_report(y_test, y_pred, target_names = iris.target_names))
```

### Improve Classifier
간단하지만 현업에서도 자주 쓰는 방식이 n_estimator을 늘리는 것임. 

```python
rfc2 = RandomForestClassifier( 
    n_estimators = 200,
    # max_features = 4,
    max_depth = 5,
    random_state = 1 
)
```

```python
rfc2.fit(X_train,y_train)
y_pred2 = rfc2.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred2)
f1 = f1_score(y_test, y_pred2, average = 'weighted')
print(f'Accuracy : {acc_score} \nf1_score = {f1}')
print(classification_report(y_test, y_pred2, target_names = iris.target_names))
```
